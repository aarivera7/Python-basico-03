```python
# En función del dataframe datos, es posible que le ayude realizar un .head para explorar el dataset,
# presentar los registros de la provincia de Loja y los registros de Zamora Chinchipe
datos[(datos["Provincia"] == "LOJA") | (datos["Provincia"] == "ZAMORA CHINCHIPE")].head()
```

```python
# Generar un archivo .csv que contenga únicamente las características:
# Nombre_Institucion
# Provincia 
# Canton
# Parroaquia
# Número de estudiantes
# Número de docentes

# Siempre que cumpla con la condición: el número de estudiantes debe ser mayor a 20 y menor o igual a 100
# El archivo debe tener el nombre: instituciones_0001.csv
datosP = datos[["Nombre_Institucion", "Provincia", "Canton", "Parroquia", "Numero_de_estudiantes", "Numero_de_docentes"]]

datosP[(datosP["Numero_de_estudiantes"] > 20) & (datosP["Numero_de_estudiantes"] <= 100)].to_csv("data/instituciones_0001.csv")
```

```python
# Generar un archivo .csv que contenga únicamente las características:
# Nombre_Institucion
# Provincia 
# Canton
# Parroquia
# Número de estudiantes
# Número de docentes

# Siempre que cumpla con la condición: el número de docentes debe ser mayor a 1 y menor o igual a 50
# El archivo debe tener el nombre: instituciones_0002.csv
datosP[(datosP["Numero_de_docentes"] > 1) & (datosP["Numero_de_docentes"] <= 50)].to_csv("data/instituciones_0002.csv")
```
